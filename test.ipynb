```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics
from bidi.algorithm import get_display
import arabic_reshaper
import os

def hebrew_text_to_pdf_reportlab(hebrew_text, output_pdf_path, font_path='NotoSansHebrew-Regular.ttf', font_size=14, page_orientation='P'):
    """
    Converts Hebrew text to a PDF file using ReportLab.

    Parameters:
    - hebrew_text (str): The Hebrew text to convert.
    - output_pdf_path (str): The path where the PDF will be saved.
    - font_path (str): Path to the TTF font file that supports Hebrew.
    - font_size (int): Size of the font in the PDF.
    - page_orientation (str): 'P' for Portrait or 'L' for Landscape.

    Returns:
    - None
    """
    # Check if font file exists
    if not os.path.isfile(font_path):
        raise FileNotFoundError(f"Font file not found: {font_path}")

    # Register the font
    font_name = 'HebrewFont'
    pdfmetrics.registerFont(TTFont(font_name, font_path))

    # Create a canvas
    if page_orientation.upper() == 'L':
        c = canvas.Canvas(output_pdf_path, pagesize=reportlab.lib.pagesizes.landscape(A4))
        width, height = reportlab.lib.pagesizes.landscape(A4)
    else:
        c = canvas.Canvas(output_pdf_path, pagesize=A4)
        width, height = A4

    # Set font
    c.setFont(font_name, font_size)

    # Reshape and reorder the Hebrew text
    reshaped_text = arabic_reshaper.reshape(hebrew_text)
    bidi_text = get_display(reshaped_text)

    # Define text position
    margin = 40
    text_object = c.beginText()
    text_object.setTextOrigin(width - margin, height - margin)  # Start from the top-right corner
    text_object.setLeading(font_size + 4)
    text_object.setFont(font_name, font_size)

    # Split the text into lines if necessary
    lines = bidi_text.split('\n')
    for line in lines:
        text_object.textLine(line)

    c.drawText(text_object)
    c.showPage()
    c.save()
    print(f"PDF successfully saved to {output_pdf_path}")


hebrew_sample = "שלום עולם! זהו מבחן להמרת טקסט בעברית ל-PDF."
output_path = "contract.pdf"
font_file = "David New Hebrew Regular.ttf"  # Ensure this file is in the same directory or provide the full path

hebrew_text_to_pdf_reportlab(hebrew_sample, output_path, font_path=font_file)
```
